```python
import numpy as np
import cv2
import xml.etree.ElementTree as ET
from xml.etree.ElementTree import Element, dump, ElementTree
import os
import shutil

num = 0
count = 0
night = 0
label_num = 0


dict = {0:0, 1:130, 2:300, 3:470, 4:630, 5:800}

file_num = 1
while(file_num <= 5):
    i = dict[file_num - 1]
    count = 0
    while(i < dict[file_num]):
        i += 1
        a = len(str(i))
        k = ""
        while(a < 3):
            k = k + "0"
            a += 1
        k = k + str(i)
        xml_file = f"D:\\CapstoneDesign\\작업중\\Surface_{str(file_num)}\\Surface_{k}"
        reg = r'.xml'
        xlsx_list = [file for file in os.listdir(xml_file) if file.endswith(reg)]
        xml = f"{xml_file}\\{xlsx_list[0]}"
        doc = ET.parse(xml)
        #root 노드 가져오기
        root = doc.getroot()
        image_tag = root.findall("image")

        point = []
        regress = 0
        blocks_if = 0
        for attr in root.iter("image"):
            name = attr.attrib["name"].split(".")[0]
            f = open(f"{xml_file}\\{name}.txt", 'w')
            blocks_if = 0
            for poly in attr.iter("polygon"):
                label = poly.attrib["label"]
                points = poly.attrib["points"].replace(";", ",").split(",")
                point = []
                if label == "sidewalk":
                    if poly.find("attribute").text == "blocks":
                        blocks_if = 1
                        f.write("blocks")
                        for num, res in enumerate(points):
                            if num % 2 == 0:
                                regress = float(res)/1920.00
                            else:
                                regress = float(res)/1080.00
                            point.append(regress)
                        for p in point:
                            f.write(f" {str(p)}")
                        f.write("\n")
                    elif poly.find("attribute").text == "damaged":
                        blocks_if = 2
                        f.write("damaged")
                        for num, res in enumerate(points):
                            if num % 2 == 0:
                                regress = float(res)/1920.00
                            else:
                                regress = float(res)/1080.00
                            point.append(regress)
                        for p in point:
                            f.write(f" {str(p)}")
                        f.write("\n")
                    count += 1
            f.close()
            if blocks_if == 0:
                os.remove(f"{xml_file}\\{name}.txt")
                count -= 1
    print(f"total {file_num}: {count}")
    file_num += 1
```
